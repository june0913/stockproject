```python
import yfinance as yf
import pandas as pd
import time

# sp500_companies.csv 파일에서 S&P 500 회사들의 티커 심볼 불러오기
sp500 = pd.read_csv('./sp500_companies.csv')
sp500_tg = sp500['Symbol']

request_count = 0
for ticker_symbol in sp500_tg:
    if request_count >= 5:
        print("Reached 5 requests, sleeping for 10 seconds...")
        time.sleep(10)  # 5개 요청 후 10초간 대기
        request_count = 0  # 요청 카운트 초기화

    # 각 주식 정보 가져오기
    try:
        stock = yf.Ticker(ticker_symbol)
        info = stock.info

        # 정보의 일부 출력
        print(f"Ticker: {ticker_symbol}")
        print(f"Company Name: {info.get('longName')}")
        print(f"Sector: {info.get('sector')}")
        print(f"Industry: {info.get('industry')}")
        print(f"Full Time Employees: {info.get('fullTimeEmployees')}")
        print("-" * 50)  # 정보 구분을 위한 구분선

    except Exception as e:
        print(f"Error fetching data for {ticker_symbol}: {e}")
    
    request_count += 1  # 요청 카운트 증가
```

```python
import mysql.connector
import pandas as pd
import yfinance as yf

# 데이터베이스 연결 설정
db_config = {
    'user': 'root',
    'password': '43qjsrnfjddl!@',
    'host': 'localhost',
    'database': 'sp500_db'
}

# S&P 500 회사들의 티커 심볼 불러오기
sp500 = pd.read_csv('./sp500_companies.csv')
sp500_tg = sp500['Symbol']

def save_stock_info_to_db():
    # 데이터베이스 연결
    cnx = mysql.connector.connect(**db_config)
    cursor = cnx.cursor()

    for ticker_symbol in sp500_tg:
        stock = yf.Ticker(ticker_symbol)
        info = stock.info
        
        # 필요한 정보를 변수에 저장
        data = {
            'symbol': ticker_symbol,
            'marketCap': info.get('marketCap'),
            'enterpriseValue': info.get('enterpriseValue'),
            'trailingPE': info.get('trailingPE'),
            'forwardPE': info.get('forwardPE'),
            'dividendYield': info.get('dividendYield'),
            'profitMargins': info.get('profitMargins'),
        }
        
        # 데이터베이스에 데이터 삽입
        query = ("""
        INSERT INTO stock_info 
        (symbol, marketCap, enterpriseValue, trailingPE, forwardPE, dividendYield, profitMargins) 
        VALUES (%(symbol)s, %(marketCap)s, %(enterpriseValue)s, %(trailingPE)s, %(forwardPE)s, %(dividendYield)s, %(profitMargins)s) 
        ON DUPLICATE KEY UPDATE 
        marketCap=VALUES(marketCap), 
        enterpriseValue=VALUES(enterpriseValue), 
        trailingPE=VALUES(trailingPE), 
        forwardPE=VALUES(forwardPE), 
        dividendYield=VALUES(dividendYield), 
        profitMargins=VALUES(profitMargins);
        """)
        
        cursor.execute(query, data)
        cnx.commit()
    
    # 연결 종료
    cursor.close()
    cnx.close()

# 함수 호출
save_stock_info_to_db()
```

```python
import mysql.connector
import pandas as pd
import yfinance as yf

# 데이터베이스 연결 설정
db_config = {
    'user': 'root',
    'password': '43qjsrnfjddl!@',
    'host': 'localhost',
    'database': 'sp500_db'
}

# 데이터베이스 연결
cnx = mysql.connector.connect(**db_config)
cursor = cnx.cursor()

# S&P 500 회사들의 티커 심볼 불러오기
sp500 = pd.read_csv('./sp500_companies.csv')

def save_data_to_db():
    for ticker_symbol in sp500['Symbol']:
        stock = yf.Ticker(ticker_symbol)
        info = stock.info
        
        # companies 테이블에 데이터 삽입 또는 업데이트
        cursor.execute("""
        INSERT INTO companies (symbol, company_name, industry, sector) 
        VALUES (%s, %s, %s, %s) 
        ON DUPLICATE KEY UPDATE 
        company_name=VALUES(company_name), 
        industry=VALUES(industry), 
        sector=VALUES(sector)
        """, (ticker_symbol, info.get('longName'), info.get('industry'), info.get('sector')))
        
        # enterprise_data 테이블에 데이터 삽입 또는 업데이트
        cursor.execute("""
        INSERT INTO enterprise_data 
        (symbol, company_name, industry, sector, netIncomeToCommon, totalRevenue, ebitda, totalDebt, currentRatio, returnOnAssets, returnOnEquity, grossMargins, operatingMargins) 
        VALUES (%s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s) 
        ON DUPLICATE KEY UPDATE 
        company_name=VALUES(company_name), 
        industry=VALUES(industry), 
        sector=VALUES(sector), 
        netIncomeToCommon=VALUES(netIncomeToCommon), 
        totalRevenue=VALUES(totalRevenue), 
        ebitda=VALUES(ebitda), 
        totalDebt=VALUES(totalDebt), 
        currentRatio=VALUES(currentRatio), 
        returnOnAssets=VALUES(returnOnAssets), 
        returnOnEquity=VALUES(returnOnEquity), 
        grossMargins=VALUES(grossMargins), 
        operatingMargins=VALUES(operatingMargins)
        """, (
            ticker_symbol, 
            info.get('longName'), 
            info.get('industry'), 
            info.get('sector'), 
            info.get('netIncomeToCommon', 0), 
            info.get('totalRevenue', 0), 
            info.get('ebitda', 0), 
            info.get('totalDebt', 0), 
            info.get('currentRatio', 0.0), 
            info.get('returnOnAssets', 0.0), 
            info.get('returnOnEquity', 0.0), 
            info.get('grossMargins', 0.0), 
            info.get('operatingMargins', 0.0)
        ))

        # 데이터베이스에 변경사항 커밋
        cnx.commit()

# 함수 호출하여 데이터 저장
save_data_to_db()

# 데이터베이스 연결 종료
cursor.close()
cnx.close()
```
