```python
import mysql.connector
import pandas as pd
import yfinance as yf

#   
db_config = {
    'user': 'root',
    'password': '1234',
    'host': 'localhost',
    'database': 'sp500_db'
}

# S&P 500    
sp500 = pd.read_csv('./sp500_companies.csv')
sp500_tg = sp500['Symbol']

def save_stock_info_to_db():
    #  
    cnx = mysql.connector.connect(**db_config)
    cursor = cnx.cursor()

    for ticker_symbol in sp500_tg:
        stock = yf.Ticker(ticker_symbol)
        info = stock.info
        
        #    
        data = {
            'symbol': ticker_symbol,
            'marketCap': info.get('marketCap'),
            'enterpriseValue': info.get('enterpriseValue'),
            'trailingPE': info.get('trailingPE'),
            'forwardPE': info.get('forwardPE'),
            'dividendYield': info.get('dividendYield'),
            'profitMargins': info.get('profitMargins'),
        }
        
        #   
        query = ("""
        INSERT INTO stock_info 
        (symbol, marketCap, enterpriseValue, trailingPE, forwardPE, dividendYield, profitMargins) 
        VALUES (%(symbol)s, %(marketCap)s, %(enterpriseValue)s, %(trailingPE)s, %(forwardPE)s, %(dividendYield)s, %(profitMargins)s) 
        ON DUPLICATE KEY UPDATE 
        marketCap=VALUES(marketCap), 
        enterpriseValue=VALUES(enterpriseValue), 
        trailingPE=VALUES(trailingPE), 
        forwardPE=VALUES(forwardPE), 
        dividendYield=VALUES(dividendYield), 
        profitMargins=VALUES(profitMargins);
        """)
        
        cursor.execute(query, data)
        cnx.commit()
    
    #  
    cursor.close()
    cnx.close()

#  
save_stock_info_to_db()
```

```python
import mysql.connector
import yfinance as yf
import pandas as pd
import time

#   
db_config = {
    'user': 'root',
    'password': '1234',
    'host': 'localhost',
    'database': 'sp500_db'
}


# MySQL  
conn = mysql.connector.connect(**db_config)
cursor = conn.cursor()

# S&P 500    
sp500 = pd.read_csv('./sp500_companies.csv')
sp500_tg = sp500['Symbol']

#    
request_count = 0
request_limit = 5
sleep_duration = 10  # 10 

#        
for ticker_symbol in sp500_tg:
    # API       
    if request_count >= request_limit:
        print(f"Reached {request_limit} requests, sleeping for {sleep_duration} seconds...")
        time.sleep(sleep_duration)
        request_count = 0

    try:
        # 1   
        hist = yf.download(ticker_symbol, period="1mo", interval="1h")

        #   
        for index, row in hist.iterrows():
            cursor.execute("""
            INSERT INTO stock_prices (symbol, date, open, high, low, close, volume) 
            VALUES (%s, %s, %s, %s, %s, %s, %s)
            """, (ticker_symbol, index.strftime('%Y-%m-%d %H:%M:%S'), 
                  row['Open'], row['High'], row['Low'], row['Close'], row['Volume']))
        conn.commit()

    except Exception as e:
        print(f"Error fetching data for {ticker_symbol}: {e}")

    request_count += 1

#   
cursor.close()
conn.close()
```

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import mysql.connector
from urllib.parse import quote_plus
import feedparser
from datetime import datetime, timedelta
import time
from email.utils import parsedate_to_datetime
from collections import defaultdict

# MySQL   
db_config = {
    'user': 'root',
    'password': '1234',
    'host': 'localhost',
    'database': 'sp500_db'
}

#     
sp500_companies = pd.read_csv('sp500_companies.csv')

def fetch_articles_rss(company_name, symbol):
    print(f"Fetching articles for {company_name}...")
    encoded_company_name = quote_plus(company_name)
    rss_url = f"https://news.google.com/rss/search?q={encoded_company_name}"
    feed = feedparser.parse(rss_url)
    
    #       
    articles_per_day = defaultdict(list)
    
    for entry in feed.entries:
        published_date = parsedate_to_datetime(entry.published).date()
        if len(articles_per_day[published_date]) < 40:
            articles_per_day[published_date].append({
                'symbol': symbol,
                'publishedAt': entry.published,
                'title': entry.title,
                'description': entry.summary,
                'url': entry.link
            })

    #        
    articles = [article for day_articles in articles_per_day.values() for article in day_articles]
    return articles

def convert_to_mysql_datetime(rfc2822_string):
    dt = parsedate_to_datetime(rfc2822_string)
    return dt.strftime('%Y-%m-%d %H:%M:%S')

def save_articles_to_db(articles):
    connection = mysql.connector.connect(**db_config)
    cursor = connection.cursor()
    
    add_article = ("INSERT INTO news_articles "
                   "(symbol, publishedAt, title, description, url) "
                   "VALUES (%s, %s, %s, %s, %s)")
    
    for article in articles:
        mysql_datetime = convert_to_mysql_datetime(article['publishedAt'])
        try:
            cursor.execute(add_article, (article['symbol'], mysql_datetime, article['title'], article['description'], article['url']))
            connection.commit()
        except mysql.connector.Error as err:
            print(f"Failed to insert article: {err}")
            connection.rollback()

    cursor.close()
    connection.close()

#    
def print_progress(index, total):
    progress = (index + 1) / total * 100
    print(f"Progress: {index + 1}/{total} ({progress:.2f}%)")

#       
total_companies = len(sp500_companies)
for index, row in sp500_companies.iterrows():
    company_name = row['Company']
    symbol = row['Symbol']
    articles = fetch_articles_rss(company_name, symbol)
    save_articles_to_db(articles)
    print_progress(index, total_companies)
    time.sleep(10)  # 10    

print("All articles have been fetched and saved.")
```
